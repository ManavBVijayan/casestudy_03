# 2. Build a regression model to predict the profit

```python
data1=data.drop(['Date'],axis=1)
```

```python
x=data1.drop(['Profit(in Thousands)'],axis=1)
y=data1['Profit(in Thousands)']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.25)
from sklearn import linear_model
lr = linear_model.LinearRegression()
model=lr.fit(x_train,y_train)
prediction=model.predict(x_test)
prediction
```

```python
from sklearn.metrics import mean_squared_error
print("Mean squared error is:",mean_squared_error(y_test,prediction))
```

from this we can see the error very small and we can use the line to predict the profit.
